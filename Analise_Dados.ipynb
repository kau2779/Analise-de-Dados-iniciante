```python
new_columns_name = {
    'member_id' : 'ID_de_Membro',
    'visit_date' : 'Data_da_Visita',
    'age' : 'Idade', 
    'gender' : 'Genêro', 
    'membership_type' : 'Tipo_Cartão_Membro',
    'workout_type' : 'Tipo_Treino',
    'workout_duration_minutes': 'Tempo_duração_Treino',
    'calories_burned' : 'Calorias_Queimadas',
    'check_in_time' : 'Horário_Entrada',
    'attendance_status' : 'Status_Presença'
}
df.rename(columns = new_columns_name, inplace = True)

df.columns
```

```python
mapeamento_genero = {
    "Female" : "Mulher", 
    "Other" : "Outro",
    "Male" : "Homem"
}

df["Genêro"] = df["Genêro"].replace(mapeamento_genero)
```

```python
mapeamento_Tipo_Cartão_Membro = {
    "Annual" : "Anual",
    "Monthly" : "Mensal",
    "Quarterly" : "Trimestal"
}

df["Tipo_Cartão_Membro"] = df["Tipo_Cartão_Membro"].replace(mapeamento_Tipo_Cartão_Membro)
```

```python
mapeamento_Tipo_Treino = {
    "Strength Training" : "Treino_Força",
    "Cardio" : "Cardio",        
    "HIIT" : "HIIT",      
    "CrossFit" : "CrosFit",
    "Yoga" : "Yoga"
}

df["Tipo_Treino"] = df["Tipo_Treino"].replace(mapeamento_Tipo_Treino)
```

```python
mapeamento_Status_Presença = {
    "Absent" : "Falta",
    "Present" : "Presente"
}

df["Status_Presença"] = df["Status_Presença"].replace(mapeamento_Status_Presença)
```

```python
#Contar a quantidade de Genêros
genero_contagem = df["Genêro"].value_counts().reset_index()
genero_contagem.columns = ["tipo_genero", "quantidade"]

#Criar Gráfico
Graf_piz = px.pie(
   genero_contagem,
   names = "tipo_genero",
   values = "quantidade",
   title = "Proporção de gênero na academia",
   hole = 0.5
)

Graf_piz.update_traces(textinfo = "percent + label")
```

```python
#Contar a idade
idade_contagem = df["Idade"].value_counts().reset_index()
idade_contagem.columns = ["Idade", "quantidade"]
#Criar o gáfico
Graf_bar = px.bar(
   idade_contagem,
   x = "Idade",
   y =  "quantidade",
   labels = {"quantidade" : "Somatória das pessoas com idades iguas"},
   title = "Idade dentro da Academia"
)
Graf_bar.update_traces(texttemplate = "%{y}", textangle = 0)
Graf_bar.show()
```

```python
#Contar o tipo de treino e as pessoas
treino_contagem = df["Tipo_Treino"].value_counts().reset_index()
treino_contagem.columns = ["Tipo de Treino", "Quantidade de pessoas"]

#Criar o gáfico
Graf_bar2 = px.bar(
    treino_contagem,
    x = "Tipo de Treino",
    y = "Quantidade de pessoas",
    title = "Modalidades de treinos na academia"
)

Graf_bar2.update_traces(texttemplate = "%{y}", textangle = 0)
Graf_bar2.show()
```

```python
#Contagem de Presença
presenca_contagem = df["Status_Presença"].value_counts().reset_index()
presenca_contagem.columns  = ["Presença", "Quantidade"]

#Criar Gráfico
Graf_piz2 = px.pie(
    presenca_contagem,
    names = "Presença",
    values = "Quantidade",
    title = "Quantidade de Presenças na Academia",
    hole = 0.5
)

Graf_piz2.update_traces(textinfo = "percent + label")
```
